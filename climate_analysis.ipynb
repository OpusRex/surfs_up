# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
engine = create_engine("sqlite:///hawaii.sqlite")
print(engine)
```

```python
# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# We can view all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

# Exploratory Climate Analysis

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results. 
#Starting from the last data point in the database. 

# Calculate the date one year from the last date in data set.
prev_year = dt.date(2017, 8, 23) - dt.timedelta(days=365)
print(prev_year)
```

```python
# Perform a query to retrieve the data and precipitation scores
results = []
results = session.query(Measurement.date, Measurement.prcp).filter(Measurement.date >= prev_year).all()
print(results)
```

```python
# Save the query results as a Pandas DataFrame and set the index to the date column
df = pd.DataFrame(results, columns=['date', 'precipitation'])
df.head()
```

```python
#set new index based on date, overwriting the original dataframe
df.set_index(df['date'],inplace=True)
df.head(10)
```

```python
#remove the extra date column, hides the index
print(df.to_string(index=False))
```

```python
# Sort the dataframe by date and print
df = df.sort_index()
print(df.to_string(index=False))
```

```python
# Use Pandas to calcualte the summary statistics for the precipitation data
df.describe()
```

```python
# Use Pandas Plotting with Matplotlib to plot the data
df.plot()
plt.tight_layout()
plt.xticks(rotation=90)
```

```python
# How many stations are available in this dataset?
session.query(func.count(Station.station)).all()
```

```python
# What are the most active stations?
# List the stations and the counts in descending order.
session.query(Measurement.station, func.count(Measurement.station)).\
    group_by(Measurement.station).order_by(func.count(Measurement.station).desc()).all()
```

```python
# Using the station id from the previous query, calculate the lowest temperature recorded, 
# highest temperature recorded, and average temperature most active station?
session.query(func.min(Measurement.tobs), func.max(Measurement.tobs), func.avg(Measurement.tobs)).\
    filter(Measurement.station == 'USC00519281').all()
```

```python
# Choose the station with the highest number of temperature observations.
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram
results = session.query(Measurement.tobs).\
filter(Measurement.station =='USC00519281').\
filter(Measurement.date >= prev_year).all()
print(results)

df = pd.DataFrame(results, columns=['tobs'])
df.head()
```

```python
df.plot.hist(bins=12)
plt.tight_layout()
```

```python
# query db and return rainfall totals from all stations
def rain_totals(month):
    
    #get proper string month name for graph display
    if month == '01':
        monthof = 'January'
    elif month == '02':
        monthof = 'Feburary'
    elif month == '03':
        monthof = 'March'
    elif month == '04':
        monthof = 'April'
    elif month == '05':
        monthof = 'May'
    elif month == '06':
        monthof = 'June'
    elif month == '07':
        monthof = 'July'
    elif month == '08':
        monthof = 'August'
    elif month == '09':
        monthof = 'September'
    elif month == '10':
        monthof = 'October'
    elif month == '11':
        monthof = 'November'
    elif month == '12':
        monthof = 'December'
        
    results = session.query(Measurement.date,Measurement.station, Measurement.prcp, func.strftime("%m", Measurement.date),func.strftime("%Y", Measurement.date)).\
                    filter(func.strftime("%m", Measurement.date)==month).all()
    #dataframe using Sqlite rain data
    df_raintemp = pd.DataFrame(results, columns=['date','station_id','precipitation','Month', 'Year'])
    
    #sort data for graph display
    df_raintemp=df_raintemp.sort_values('date',axis=0)
   
    #create graph 
    
    #get string of month actual name
    str=f'Rain totals for all stations for {monthof} (all years)'
                
    X=df_raintemp.plot(x='date',y='precipitation')
    plt.tight_layout()                                          
    plt.xticks(rotation=90)
    plt.xlabel('date')
    plt.ylabel ('Rainfall(Inches)')
    plt.title(str)
    return df_raintemp.describe(), "           ", df_raintemp.head(), "      ", X
```

```python
# query db and return tempertures from all stations
def temps_measured(month):
    
    #get proper string month name for graph display
    if month == '01':
        monthof = 'January'
    elif month == '02':
        monthof = 'Feburary'
    elif month == '03':
        monthof = 'March'
    elif month == '04':
        monthof = 'April'
    elif month == '05':
        monthof = 'May'
    elif month == '06':
        monthof = 'June'
    elif month == '07':
        monthof = 'July'
    elif month == '08':
        monthof = 'August'
    elif month == '09':
        monthof = 'September'
    elif month == '10':
        monthof = 'October'
    elif month == '11':
        monthof = 'November'
    elif month == '12':
        monthof = 'December'
    
    results = session.query(Measurement.date,Measurement.station, Measurement.tobs, func.strftime("%m", Measurement.date),func.strftime("%Y", Measurement.date)).\
                    filter(func.strftime("%m", Measurement.date)==month).all()
    #dataframe using Sqlite temperatre data
    df = pd.DataFrame(results, columns=['date','station_id','Temperature','Month', 'Year'])

    # create bar graph                                      
    str=f'Temperature of all stations for {monthof} (all years)'           
    y= df.plot.hist(bins=12)
    plt.tight_layout()
    plt.xlabel('Temperature(F)')
    plt.ylabel ('Frequency(days) for month')
    plt.title(str)
    
    return df.describe(),"            ",df.head(),"           ",y
```
